#### Basic Syntax and Semantics

```python
# Single Line Comment
'''
Welcome to the Python - Multi line comments
'''

print("Let's Start learning Python!!!")
```

```python
# Basis syntax Rules in Python
# Python is case sensitive
name="Mohammad"
Name="Tanvir"
print(name)
print(Name)
```

```python
# Indentation
# Python uses uses indentation to define blocks of code.
# Consistent use of spaces (commonly 4) or a tab is required

age=32
if age>32:
    print(age)
print(age)
```

```python
# Line Continuation
total= 1+2+3+4+5+6+\
    7+8+9
print(total)
```

```python
# Multiple Statements in single line
x=5;y=10;z=x+y
print(z)
```

```python
## Understand Sematics in Python
# Variable assignment
age = 32 ## age is interger
name = "Tanvir" ## name is a string
print(type(age))
print(type(name))
```

```python
# Type Inference
# Python Infers type of the variable at the runtime

var_int = 10
print(type(var_int))
var_str = "Tanvir"
print(type(var_str))
```

```python
# Code example of Indentation

if True:
    print("Correct Indentation!!!")
    if False:
        print("This won't print")
    print("This will print")
print("Outside the print block!!!")
```
